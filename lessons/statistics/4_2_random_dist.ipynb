Think Stats Chapter 4 Exercise 2 (a random distribution)
Exercise 2   The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.
Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

```python
import thinkstats2
from random import random
import matplotlib.pyplot as plt
import thinkplot
```

```python
dist = [random() for i in range(1000)]
pmf = thinkstats2.Pmf(dist, 'random') 
```

```python
thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='Value', ylabel='Pmf', 
               title='Random numbers')
```

```python
cdf = thinkstats2.Cdf(dist, label='random') 
```

```python
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='Value', ylabel='Cdf', 
               title='Random numbers')
```

Yes, the distribution is uniform.
